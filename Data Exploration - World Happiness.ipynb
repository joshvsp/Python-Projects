```python
#Which are the lowest AND highest ranked countries in 2015? 

low_2015 = wh[wh['Year'] == 2015][['Year', 'Country', 'Happiness Score']].sort_values(by='Happiness Score')[:10]
high_2015 = wh[wh['Year'] == 2015][['Year', 'Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False)[:10]


# Sorting values so higher scores are at the top
low_2015 = low_2015.sort_values(by='Happiness Score', ascending=False)

low_high_2015 = pd.concat([low_2015,high_2015]).sort_values(by='Happiness Score', ascending=False)

sns.barplot(x='Happiness Score', y='Country', data=low_high_2015 ,orient = 'h')
plt.title('Highest vs. Lowest Ranked Happiness Scores (2015)')
plt.tight_layout()
```

```python
#Which are the lowest AND highest ranked countries in 2016? 

low_2016 = wh[wh['Year']==2016][['Year', 'Country', 'Happiness Score']].sort_values(by='Happiness Score')[:10]
high_2016 = wh[wh['Year']==2016][['Year', 'Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False)[:10]

low_2016 = low_2016.sort_values(by='Happiness Score', ascending=False)

low_high_2016 = pd.concat([high_2016,low_2016])

sns.barplot(x='Happiness Score', y='Country', data=low_high_2016, orient='h')
```

```python
#Comparing Top Scores between 2015 and 2016 

highs = pd.concat([high_2015,high_2016])
sns.barplot(x='Happiness Score', y='Country', hue='Year', data=highs, orient='h')
```

```python
#Which countries experienced the greatest change in happiness rank from 2015 to 2016? 


#Only including countries present within both years
wh2015 = wh[wh['Year'] == 2015]
wh2016 = wh[wh['Year'] == 2016]

wh_delta = pd.merge(wh2015, wh2016, how='inner', on=['Country'])

#Sorting alphabetically

wh_delta.sort_values(by='Country', inplace=True)

#Renaming columns

wh_delta.rename(columns={'Happiness Rank_x':'Happiness Rank 2015', 'Happiness Rank_y':'Happiness Rank 2016'}, inplace=True)

#Creating 'Rank Change' column

wh_delta['Rank Difference'] = wh_delta['Happiness Rank 2015'] - wh_delta['Happiness Rank 2016']
```

```python
#Creating a table visualizing rank difference

#Filtering companies that shifted more 10 ranks
diff_ten = wh_delta[(wh_delta['Rank Difference'] <= -10) | (wh_delta['Rank Difference'] >= 10)] 

sns.barplot(x='Rank Difference', y='Country', data=diff_ten, orient='h')

plt.tight_layout()
plt.title('Rank Differences')
```

```python
#How much did these countries scores change?

sns.barplot(x=(diff_ten['Happiness Score_y'] - diff_ten['Happiness Score_x']), y=diff_ten['Country'], orient='h')

plt.tight_layout()
plt.title('Score Difference')
```

```python
#How about regionally?

#Creating new Score Difference column
wh_delta['Score Difference'] = wh_delta['Happiness Score_y'] - wh_delta['Happiness Score_x']

#Grouping by region
region = wh_delta.groupby('Region_x')
```

```python
#Creating bar graph of average score difference 

region['Score Difference'].mean().plot.barh()
plt.title('Score difference from 2015 to 2016')
```
